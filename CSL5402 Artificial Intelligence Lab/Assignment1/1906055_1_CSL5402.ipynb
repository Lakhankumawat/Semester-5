1.WAP to implement a simple calculator. The output should be something like:Select operation.

```python
# -*- coding: utf-8 -*-

def calculator(num1,num2,chosen):
    if chosen==1:
        print(str(num1)+"+"+str(num2)+" =",num1+num2)
    elif chosen==2:
        print(str(num1)+"-"+str(num2)+" =",num1-num2)
    elif chosen==3:
        print(str(num1)+"x"+str(num2)+" =",num1*num2)
    elif chosen==4:
        print(str(num1)+"/"+str(num2)+" =",num1/num2)
    else:
        print("invalid choice")
choice=int(input("Enter choice (1,2,3,4):"))
num1=int(input("Enter first Number :"))
num2=int(input("Enter second Number :"))
calculator(num1,num2,choice)
```

2.WAP to check if a year is leap. Use % operator



```python

year=int(input("Enter Year :"))

if year%4==0:
    if year%100==0:
        if year%400==0:
            print("Leap year")
        else:
            print("Not a leap year")   
    else:
        print("Leap year")

else:
    print("Not a leap year")
```

3.WAP to find factorial of a number. Use range() function. 

```python

num=int(input("Enter Number :"))
fact=1
s=''
for i in range(1,num+1):
    fact*=i
    if i!=num:
        s+=str(i)+"x"
    else:
        s+=str(i)+"="

print(s,fact)
```

```python
# -*- coding: utf-8 -*-
import random
suites = ['Hearts', 'Diamonds', 'Clubs', 'Spades']
count=random.randint(1,13)
suite=random.choice(suites)
print(str(count)+" of",suite)
```
